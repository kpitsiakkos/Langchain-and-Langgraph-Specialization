```python
import os

# Option A: Load from file locally (add hf_token.txt to .gitignore)
with open("hf_token.txt", "r") as file:
    hf_token = file.read().strip()

# Store in environment variable (safer way to pass it around)
os.environ["HF_TOKEN"] = hf_token

print("Token loaded successfully")
```

```python
from huggingface_hub import InferenceClient

# Always read from environment variable, never hardcode the token
hf_token = os.environ.get("HF_TOKEN")

client = InferenceClient(
    provider="featherless-ai",
    api_key=hf_token,
)

print("Client ready")
```

```python
our_query = "What is the currency of India?"

completion = client.chat.completions.create(
    model="mistralai/Mistral-7B-Instruct-v0.2",
    messages=[
        {"role": "user", "content": our_query}
    ],
    max_tokens=50,
)

print(completion.choices[0].message.content)
```
