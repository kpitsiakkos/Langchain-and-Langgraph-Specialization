<h2 style="color: #27AE60;"> Initialize Embeddings</h2>

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings()
```

<h2 style="color: #E67E22;"> Get Embedding for Search Word</h2>

```python
import pandas as pd
df = pd.read_excel('data.xlsx')
print(df)

def get_embedding(word):
    return embeddings.embed_query(word)

df['embedding'] = df['Words'].apply(get_embedding)
df.to_csv('word_embeddings.csv', index=False)
```

<h2 style="color: #E67E22;"> Calculate Similarity Scores</h2>

```python
our_Text = "Mango"
text_embedding = embeddings.embed_query(our_Text)
print(f"Our embedding is {text_embedding}")
```

<h2 style="color: #8E44AD;"> Results</h2>

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

df["similarity score"] = df['embedding'].apply(
    lambda x: cosine_similarity([x], [text_embedding])[0][0]
)
df.sort_values("similarity score", ascending=False).head(10)
```
