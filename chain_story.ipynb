```python
import os
from dotenv import find_dotenv, load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate

load_dotenv(find_dotenv())

template = """ 
 As a children's book writer, please come up with a simple and short (90 words)
 lullaby based on the location {location}
 and the main character {name}
 STORY:
"""

prompt = PromptTemplate(input_variables=["location", "name"], template=template)
llm = ChatOpenAI(temperature=0.7, model="gpt-3.5-turbo")

# LLMChain is gone in v1.x — use LCEL pipe syntax instead
chain = prompt | llm
result = chain.invoke({"location": "Zanzibar", "name": "Maya"})
print(result.content)
```
